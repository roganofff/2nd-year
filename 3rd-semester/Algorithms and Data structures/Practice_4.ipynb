# Задача 1

Напишите функцию возврщающую n-ное число Фибоначчи.

```python
def fib(n):
    if n == 0: return 0
    if n == 1: return 1
    return fib(n - 2) + fib(n - 1)


d = {0: 0, 1: 1}

def fib2(n):
    if n in d.keys():
        return d[n]
    else:
        fn = fib2(n - 2) + fib2(n - 1)
        d[n] = fn
        return fn
    
def fib3(n):
    d = {0: 0, 1: 1}
    for i in range(2, n + 1):
        d[i] = d[i - 1] + d[i - 2]
    
    return d[n]


fib3(1000000)
```

Определите минимальный и максимальный расход энергии при переходе робота в правую верхнюю клетку поля. В ответе запишите два числа: сначала минимальный расход энергии, затем – максимальный.
Исходные данные записаны в файле 3.txt. 

```python
f = [list(map(int, s.split())) for s in open('3.txt').readlines()]

n = len(f)
m = len(f[0])

en = [[0] * m for _ in range(n)]
en[-1][0] = f[-1][0]

for i in range(1, m):
    en[-1][i] = abs(f[-1][i - 1] - f[-1][i]) + en[-1][i - 1]

for i in range(n - 2, -1, -1):
    en[i][0] = abs(f[i][0] - f[i + 1][0]) + en[i + 1][0]

for i in range(n - 2, -1, -1):
    for j in range(1, m):
        en[i][j] = min(abs(f[i][j] - f[i][j - 1]) + en[i][j - 1], 
                       abs(f[i][j] - f[i + 1][j]) + en[i + 1][j])

for s in en:
    print(*s)

en
```

# Задача 5

Даны две последовательности, требуется найти длину их наибольшей общей подпоследовательности.

```python
def t5(a: str, b:str) -> int:
    n, m = len(a), len(b)

    dp = [[0] * m for _ in range(n)]

    for i in range(n):
        for j in range(m):
            if a[i] == b[j]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp
t5('12345', '27325')
```
